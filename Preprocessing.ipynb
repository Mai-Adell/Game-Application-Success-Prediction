```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
import datetime as dt
from sklearn import metrics
import itertools
import functools
import regex as re
data = pd.read_csv('games-regression-dataset.csv') 
data.shape
```

```python

#drop first url as the diferent parts of it is the same as the next 2 columns 

data.drop(['URL'],axis=1,inplace=True)
data.drop(['Description'],axis=1,inplace=True)


##print (data.head)
```

```python
x=data
x['Icon URL']=x['Icon URL'].str.replace("https://is5-ssl.mzstatic.com/image/thumb/Purple",'')
x['Icon URL']=x['Icon URL'].str.replace("https://is4-ssl.mzstatic.com/image/thumb/Purple",'')
x['Icon URL']=x['Icon URL'].str.replace("https://is3-ssl.mzstatic.com/image/thumb/Purple",'')
x['Icon URL']=x['Icon URL'].str.replace("https://is2-ssl.mzstatic.com/image/thumb/Purple",'')
x['Icon URL']=x['Icon URL'].str.replace("https://is1-ssl.mzstatic.com/image/thumb/Purple",'')
x['Icon URL']=x['Icon URL'].str.replace("/source/512x512bb.jpg",'')



```

```python
s=x[['Icon URL','Name']]

sa=pd.get_dummies(x,columns=x['Icon URL'],drop_first=True)
sa
```

```python
labelEncoder =LabelEncoder()
x['Name']=labelEncoder.fit_transform(x['Name'])
x['Icon URL']=labelEncoder.fit_transform(x['Icon URL'])
x['Subtitle']=labelEncoder.fit_transform(x['Subtitle'])
x['Developer']=labelEncoder.fit_transform(x['Developer'])
x['Languages']=labelEncoder.fit_transform(x['Languages'])
x['Primary Genre']=labelEncoder.fit_transform(x['Primary Genre'])
x['Genres']=labelEncoder.fit_transform(x['Genres'])
x.head()
```

```python


x['Original Release Date']=pd.to_datetime(x['Original Release Date'],dayfirst=True)
x['Original Release Date']=x['Original Release Date'].map(dt.datetime.toordinal)
x['Current Version Release Date']=pd.to_datetime(x['Current Version Release Date'],dayfirst=True)
x['Current Version Release Date']=x['Current Version Release Date'].map(dt.datetime.toordinal)
x.dtypes
```
